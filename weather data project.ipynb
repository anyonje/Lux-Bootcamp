1. Find all records where the weather was exactly clear

```python
clear_weather_df = weather_df[weather_df['Weather'] == 'Clear']
clear_weather_df
```

2. Find the number of times the wind speed was exactly 4 km/hr.

```python
wind_speed_df = weather_df[weather_df['Wind Speed_km/h'] == 4]
num_of_times = len(wind_speed_df)
print("Number of times:", num_of_times)
wind_speed_df 
```

4. Rename the column "Weather" to "Weather_Condition."

```python
new_column_name = 'Weather_Condition'
weather_df.rename(columns={'Weather': new_column_name}, inplace=True)
weather_df.head()
```

5. What is the mean visibility of the dataset?

```python
mean_visibility = weather_df['Visibility_km'].mean()
print("Mean Visibility:", mean_visibility)
```

6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
filtered_df = weather_df[(weather_df['Wind Speed_km/h'] > 24) & (weather_df['Visibility_km'] == 25)]
num_records = len(filtered_df)
print("Number of records:", num_records)
filtered_df
```

8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40

```python
filtered_df = weather_df[(weather_df['Weather_Condition'] == 'clear') & (weather_df['Rel Hum_%'] > 50) | (weather_df['Weather_Condition'] ==  'Clear')& (weather_df['Visibility_km'] > 40)]
filtered_df
num_instances = len(filtered_df)
print("Number of instances:", num_instances)
filtered_df
```

9. Find the number of weather conditions that include snow

```python
snow_df = weather_df[(weather_df['Weather_Condition'] == 'Snow')]
snow_num = len(snow_df)
print("Number of snow conditions:", snow_num)
snow_df
```
